```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("telco.csv")
```

```python
data_filled = data.fillna('Unknown')
data_filled
```

```python
y = data_filled.Churn
features = data_filled.columns.drop( 'Churn' )
x = data_filled[features]
```

```python
#Train/Test Split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2,stratify=y)
```

```python
train_x.insert(0,train_y.name,train_y)
test_x.insert(0,test_y.name, test_y)
```

```python
train_x.to_csv('train.csv',index=0)
```

```python
test_x.to_csv('test.csv',index=0)
```
